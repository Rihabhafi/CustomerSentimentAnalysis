```python
from pyspark.sql import SparkSession

# Créer une session Spark
spark = SparkSession.builder.appName("CustomerSentimentAnalysis").getOrCreate()

# Chemin vers le fichier CSV extrait
file_path = "/content/test.csv"

# Charger le fichier CSV dans un DataFrame PySpark
df = spark.read.csv(file_path, header=True, inferSchema=True)

# Afficher les premières lignes
df.show(5)

# Vérifier le schéma
df.printSchema()
```

```python
# Convertir le DataFrame Spark en pandas
df_pandas = df.select("sentiment").toPandas()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Compter les occurrences de chaque catégorie
sentiment_counts = df_pandas["sentiment"].value_counts()

# Créer un graphique à barres
plt.figure(figsize=(8, 6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette="viridis")
plt.title("Répartition des Sentiments", fontsize=16)
plt.xlabel("Catégories de Sentiment", fontsize=14)
plt.ylabel("Nombre d'Avis", fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
# Créer un graphique circulaire
plt.figure(figsize=(8, 8))
sentiment_counts.plot.pie(autopct='%1.1f%%', startangle=140, colors=sns.color_palette("viridis", len(sentiment_counts)))
plt.title("Répartition des Sentiments (Graphique Circulaire)", fontsize=16)
plt.ylabel("")  # Supprime l'étiquette de l'axe Y
plt.show()
```

```python
# Grouper les données et compter les avis par genre et sentiment
gender_sentiment_counts = (
    df.groupBy( "sentiment")
    .count()
    .toPandas()
)

# Afficher les résultats
print(gender_sentiment_counts)
```

```python
from pyspark.sql.functions import count

# Calculer la distribution des sentiments par tranche d'âge
age_sentiment_counts = (
    df.groupBy("Age of User", "sentiment")
    .agg(count("sentiment").alias("count"))
    .orderBy("Age of User")
)

# Afficher les résultats
age_sentiment_counts.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Convertir en DataFrame Pandas pour la visualisation
age_sentiment_counts_pd = age_sentiment_counts.toPandas()

# Restructurer les données pour un graphe empilé
pivot_table = age_sentiment_counts_pd.pivot(index="Age of User", columns="sentiment", values="count").fillna(0)

# Créer un graphe empilé
pivot_table.plot(
    kind="bar",
    stacked=True,
    figsize=(10, 6),
    color=["red", "blue", "green"],
    title="Distribution des Sentiments par Tranche d'Âge"
)
plt.xlabel("Tranche d'Âge")
plt.ylabel("Nombre de Tweets")
plt.legend(title="Sentiment", labels=["Négatif", "Neutre", "Positif"])
plt.show()
```

```python
# Calculer la distribution des sentiments par tranche d'âge et pays
age_country_sentiment = (
    df.groupBy("Age of User", "Country", "sentiment")
    .count()
    .orderBy("Age of User", "Country")
)

# Afficher les résultats
age_country_sentiment.show(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Grouper les données par pays et sentiment
country_sentiment_counts = (
    df.groupBy("Country", "sentiment")
    .count()
    .orderBy("Country", "sentiment")
)

# Convertir en DataFrame Pandas pour tracer le graphe
country_sentiment_counts_pd = country_sentiment_counts.toPandas()

# Limiter le nombre de pays (si nécessaire) pour une visualisation claire
top_countries = country_sentiment_counts_pd["Country"].value_counts().index[:10]
filtered_data = country_sentiment_counts_pd[
    country_sentiment_counts_pd["Country"].isin(top_countries)
]

# Tracer le graphe
plt.figure(figsize=(14, 7))
sns.barplot(
    x="Country",
    y="count",
    hue="sentiment",
    data=filtered_data,
    palette=["red", "blue", "green"]
)

# Customisation
plt.title("Distribution des Sentiments par Pays", fontsize=16)
plt.xlabel("Pays", fontsize=12)
plt.ylabel("Nombre de Tweets", fontsize=12)
plt.legend(title="Sentiments", labels=["Négatif", "Neutre", "Positif"], fontsize=10)
plt.xticks(rotation=45, ha="right")
plt.show()


```

```python
df.printSchema()
df.select("text").show(truncate=False)
```

```python
df = df.filter(df["text"].isNotNull())
```

```python
from pyspark.ml.feature import Tokenizer, CountVectorizer
from pyspark.ml import Pipeline

# Tokenizer pour transformer le texte en mots
tokenizer = Tokenizer(inputCol="text", outputCol="words")

# CountVectorizer pour transformer les mots en vecteurs numériques
vectorizer = CountVectorizer(inputCol="words", outputCol="features")

# Construire le pipeline
pipeline = Pipeline(stages=[tokenizer, vectorizer])

# Ajuster le pipeline sur les données
model = pipeline.fit(df)

# Transformer les données
transformed_df = model.transform(df)
transformed_df.select("text", "words", "features").show(truncate=False)
```

```python
from pyspark.ml.feature import StringIndexer

indexer = StringIndexer(inputCol="sentiment", outputCol="label")
labeled_df = indexer.fit(transformed_df).transform(transformed_df)

# Vérifiez si la colonne label a été ajoutée
labeled_df.select("text", "sentiment", "label").show()
```

```python
# Diviser les données en jeu d'entraînement et de test
train_df, test_df = labeled_df.randomSplit([0.8, 0.2], seed=123)
```

```python
from pyspark.ml.classification import LogisticRegression

# Entraîner un modèle de régression logistique
lr = LogisticRegression(featuresCol="features", labelCol="label")
lr_model = lr.fit(train_df)

# Faire des prédictions sur le jeu de test
predictions = lr_model.transform(test_df)

# Afficher les prédictions
predictions.select("text", "sentiment", "prediction").show(truncate=False)
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# Évaluer la précision du modèle
evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print(f"Accuracy: {accuracy}")
```

```python
# Convertir les colonnes nécessaires en Pandas DataFrame
predictions_pandas = predictions.select("sentiment", "prediction").toPandas()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Créer des colonnes lisibles pour le sentiment réel et prédit
predictions_pandas["Predicted Sentiment"] = predictions_pandas["prediction"].replace({
    0.0: "neutral",
    1.0: "positive",
    2.0: "negative"
})

# Distribution des sentiments réels et prédits
plt.figure(figsize=(12, 6))
sns.countplot(data=predictions_pandas, x="sentiment", palette="coolwarm", label="Real Sentiments")
sns.countplot(data=predictions_pandas, x="Predicted Sentiment", palette="pastel", label="Predicted Sentiments", alpha=0.7)
plt.title("Real vs Predicted Sentiments")
plt.legend(["Real Sentiments", "Predicted Sentiments"])
plt.xlabel("Sentiments")
plt.ylabel("Count")
plt.show()
```
